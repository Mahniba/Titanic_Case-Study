### Introductory Story for the Notebook
The Titanic: A Real-World Case for Descriptive and Inferential Statistics
On the night of April 14th, 1912, the RMS Titanic—one of the largest and most luxurious ships of its time—collided with an iceberg during its maiden voyage. Of the more than 2,200 passengers and crew on board, over 1,500 tragically lost their lives.
More than a century later, the Titanic remains one of the most studied disasters in history—not only by historians, but by data scientists and statisticians. The incident produced detailed passenger records, including information about each traveler’s age, sex, ticket class, family size, fare, and survival status.
This data has become a classic dataset in the world of analytics, and it gives us a powerful opportunity to apply statistics to a real-world event.

## Why This Dataset Fits Our Course of Study

In this notebook, we use the Titanic dataset to bring together two major concepts from our statistics curriculum:
 1. Descriptive Statistics :  “What does the data look like?”
  We will summarize the dataset to understand patterns such as:
  What was the average age of passengers?
  Did more men or women survive?
  How many people travelled in each class?
  What is the distribution of fares?

Descriptive statistics helps us organize, visualize, and describe the data in meaningful ways using measures like mean, median, standard deviation, histograms, and boxplots.

 2. Inferential Statistics :  “What can we conclude about the population?”
  With inferential methods, we take a sample (the available data) and make educated generalizations or test hypotheses, such as:
  Did women truly have a higher survival rate than men?
  Can we be confident that older passengers were less likely to survive?
  What is the 95% confidence interval for the average age?
  Is there a statistically significant difference in fare between survivors and non-survivors?

Inferential statistics helps us move beyond simple observations and make scientifically valid claims using confidence intervals, t-tests, and hypothesis testing.

# Inferential Statistics


Hypothesis Test: Did Women Survive More Than Men?
Form hypotheses:
H0: Survival rates are the same for men and women
H1: Women survived at higher rates

```python
# survivors
survived_counts = df.groupby('sex')['survived'].sum()
# total passengers
total_counts = df['sex'].value_counts()

stat, pval = proportions_ztest(survived_counts, total_counts)
stat, pval
```

## Conclusion
If p-value < 0.05 → Women survived significantly more than men.

Confidence Interval for Mean Age
Estimate the average age of the entire Titanic population:

```python
import numpy as np
from scipy import stats

ages = df['age'].dropna()
mean_age = ages.mean()
std_age = ages.std()
n = len(ages)

# 95% Confidence Interval
ci = stats.t.interval(0.95, df=n-1, loc=mean_age, scale=std_age/np.sqrt(n))
print("95% Confidence Interval for mean age:", ci)


```

Chi-Square Test: Is Survival Related to Passenger Class?

```python
contingency_table = pd.crosstab(df["class"], df["survived"])
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)

chi2, p
```

# Conclusion
If p < 0.05, survival is statistically linked to passenger class
Conclusion: Class strongly affects survival (1st class survived more)

### Test for Randomness
The sequence of survival (0 = died, 1 = survived) should be random if survival was by chance.

```python
# Drop missing values from survival column
survival_seq = df['survived'].dropna()

# Runs test for randomness
z_stat, p_value = runstest_1samp(survival_seq, cutoff='mean')

print("Runs Test for Randomness (Survival)")
print("Z-statistic:", z_stat)
print("p-value:", p_value)
```

```python
if p_value < 0.05:
    print(" The sequence is NOT random (p < 0.05).")
else:
    print("The sequence appears random (p > 0.05).")
```
